# The Method of Testing the Solution.
This experiment will be a year long. Initially, I want to poll two thousand people in Western Europe randomly and ask them about their eating, sleeping, and exercise habits along with any vices (e.g. anything like smoking or drinking, and drugs if any) these participants have and how happy they are on a scale of one to seven. I chose two thousand people as my sample size in order to increase the validity of the A/B and A/A testing.  I will take the mean result from each category and set the mean result as the benchmark and alterations that I want the participants in Central/Eastern Europe to make. To be cautious, I will also look at the median and standard deviation of each metric as to make sure that the mean is the best option to use as a benchmark.

Afterwards, I want to enlist the help of one thousand participants randomly selected across Central Europe and one thousand randomly selected participants in Eastern Europe. Randomly selecting individuals will control for selection bias, and that way I will have a meaningful sample size that isn’t skewed. I realize that not all participants selected may want to participate and my solution for that is to analyze what sort of participants already said yes. If I’m missing a lot of participants from a particular age range, or ethnicity, or gender, or socioeconomic status, I will purposefully search for that missing category, but if all things are equal, I will randomly select someone until I have all the participants I need to conduct my study.  

Five hundred participants from Central Europe and five hundred participants from Eastern Europe will serve as the control group (aka Group A). At the end of the year, I will poll these individuals again halfway through the year and at the end of the year to see if anything has changed.
