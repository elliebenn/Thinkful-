```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('2015.csv')
df.columns
```

```python
df.head()
```

```python
sns.set(style='darkgrid')

sns.distplot(df['Family'])
plt.show()
```

There's a density curve to go along with this histogram which is nice but no automatically generated label.

```python
sns.kdeplot(df['Happiness Score'], shade=True)
plt.show()
```

This is a filled out curve that doesn't look too bad. Automatically generated the label too. 

```python
sns.distplot(df['Family'], kde=False, rug=True)
plt.show()
```

There are ticks for each observation in the histogram but there is no automatically generated label. It's kind of hard to generate graphs for single variables. 

```python
plt.plot(df['Happiness Score'])
plt.show()
```

A line graph that shows happiness over time. Can't really say much about it since I didn't add any context or labels to make this easy to understand.

```python
alpha = df['Freedom']
beta = df['Happiness Score']

plt.scatter(alpha, beta)
plt.show()
```

A scatter plot shows correlation between two variables and this could be better by showing the correlation value and if the labels showed waht was being compared.

```python
sns.pairplot(df)
plt.show()
```

Pairplots show the pairwise relationships between variables by either using histograms or scatterplots. It's really quick to see if there are correlations between variables, but the graphs are sort of hard to look at. There are too many things to look at so I probably won't use this often. 

```python
sns.swarmplot(x='Region', y='Happiness Score', data=df)
plt.xticks(rotation=85)
plt.show()
```

A Swarm plot shows us the different regions happiness score by points on a graph. The weakness for this graph is that it really isn't descriptive enough. For example, I don't know how many points are in each variable of the graph.

```python
sns.boxplot(x='Region', y='Happiness Score', data=df,palette='pastel')

plt.title('Plot: Happiness Score by Region')
plt.figure( figsize=(45, 45))

sns.despine(offset=10, trim=True)


plt.show()
```

A boxplot is nice in that it shows the IQR of each country in terms of happiness and it shows the extremity of each region, but there were too many regions here so it seems that it's a bit too cluttered.

```python
sns.set(style='darkgrid')

g = sns.factorplot(x='Happiness Score', y='Region', data=df,
                  size=7, kind='point', palette='pastel', ci=95, dodge=True,join=False)
g.despine(left=True)
g.set_ylabels('Tips')
g.set_xlabels('')
plt.title('Pointplot: Meal Tips by Time and Gender')
plt.show()
```

According to the lesson, pointplot has all the strengths of a barplot but none of its inherent weaknesses. It uses less space to convey the same amount of informatio nand the difference in size between groups is the comparison of the error bars. 

```python
sns.barplot(x='Region', y='Happiness Score', data=df,)
plt.title('Barplot: Region to Happiness')
plt.show()
```

Bar plots are nice in that they show the standard error at the top. However, the names are once again in the way. Lets try the horizontal bar plot next. 

```python
f, ax = plt.subplots(figsize=(6, 15))

sns.set_color_codes("pastel")
sns.barplot(x="Happiness Score", y="Region", data=df,
             color="b")

plt.show()
```

After thinking about it, this is probably the best way to show the barplot and the box plot later where they won't be as cluttered. This is much easier to look at compared to before when all the region names were jumbled together. 


```python
sns.boxplot(x='Happiness Score', y='Region', data=df,palette='pastel')

plt.title('Plot: Happiness Score by Region')
plt.figure( figsize=(45, 45))

sns.despine(offset=10, trim=True)


plt.show()
```

Here we can see a box plot that is scaled a bit better so that we can see the true difference. Compared to a histogram, this box plot doesn't have a 0 at the baseline whereas a histogram or bar plot always needs a 0 value as the baseline in order to show true length and have a comparison that isn't as skewed by size. This box plot also makes it easier for me to tell which regions are happier and the range of happiness in each region.
